```python
import queue
import pandas as pd
import re
import heapq
from heapq import *
```

```python
#Jaccadian similarity function
def jacc(m1,m2):


    union = m1.union(m2)

    size_union = union.__len__()
    size_intersection = m1.__len__() + m2.__len__() - size_union

    return size_intersection/size_union
```

```python
def SIM(i,l1,l2,C, method="SL"):
    if method=="SL":
        return min( C[i][l1][0], C[i][l2][0] )

    elif method=="CL":
        return max( C[i][l1][0], C[i][l2][0] )
```

```python
#to creat num_clusters  from the data
def Hirar_Algo_Clus(doc_list,num_clusters,method="SL"):

    num_docs = len(doc_list)
    C = [ [ [0,-1] for j in range(num_docs)] for i in range(num_docs)]

    I = [0 for i in range(num_docs)]
    pq_list = []
    cluster_list = []

    for i in range(num_docs):
        pq_i = []
        heapq.heapify(pq_i)

        doc1 = doc_list[i]
        
        for j in range(num_docs):

            if i==j:
                continue
            
            doc2 = doc_list[j]
            
            C[i][j][0] = -jacc(doc1,doc2)
            C[i][j][1] = j

    
            heapq.heappush(pq_i, C[i][j])
        
        
        I[i] = 1
        pq_list.append(pq_i)
        cl_i = [i]
        
        cluster_list.append(cl_i)
        
        
        
    A= []

    
    for _ in range(num_docs-num_clusters):

        min_val = 1

        for k in range(num_docs):
            if I[k]==0:
                continue
            else:
                if (pq_list[k][0])[0] < min_val:
                    min_val = (pq_list[k][0])[0]
                    k1 = k
                    k2 = (pq_list[k][0])[1]

        A.append((k1,k2))
        I[k2] = 0
        cluster_list[k1].extend(cluster_list[k2])
        pq_list[k1] = []
        heapq.heapify(pq_list[k1])

        for i in range(num_docs):
            if I[i]==0 or i==k1:
                continue
            
            pq_list[i].remove(C[i][k1])
            pq_list[i].remove(C[i][k2])
            heapq.heapify(pq_list[i])

            
            (C[i][k1])[0] = SIM(i,k1,k2,C,method=method)
            heapq.heappush(pq_list[i],C[i][k1])

            (C[k1][i])[0] = (C[i][k1])[0]
            heapq.heappush(pq_list[k1],C[k1][i])
        
        final_clusters  = []
        
        for i in range(num_docs):
            if I[i]==1:
                final_clusters.append(cluster_list[i])

        
    return A,final_clusters
```
